```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('/content/50_Startups.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
ct= ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])], remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
X_train.shape
```

```python
multi_regressor = LinearRegression()
multi_regressor.fit(X_train,y_train)
```

```python
y_pred = multi_regressor.predict(X_test)
```

```python
print(multi_regressor.coef_)
print(multi_regressor.intercept_)
```

Making a single prediction (for example the profit of a startup with R&D Spend = 160000, Administration Spend = 130000, Marketing Spend = 300000 and State = 'California')

```python
print(multi_regressor.predict([[1,0,0,160000,130000,300000]]))
```

```python
print('Variance score: {}'.format(multi_regressor.score(X_test, y_test)))
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(multi_regressor.predict(X_train), multi_regressor.predict(X_train)-y_train, color = "green", s = 10, label = 'Train data')
plt.scatter(multi_regressor.predict(X_test), multi_regressor.predict(X_test)-y_test, color = "blue", s = 10, label = 'Test data')
plt.hlines(y = 0, xmin = 0, xmax = 200000, linewidth = 2)
plt.legend(loc = 'upper right')
plt.title("Residual errors")
plt.xlabel("Profit")
plt.ylabel("Error")
plt.show()
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

**POLYNOMIAL REGRESSION**

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
print(X_poly.shape)
X_poly_test = poly.transform(X_test)
```

```python
poly.fit(X_poly, y_train)
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y_train)
```

```python
y_poly_pred = poly_regressor.predict(X_poly_test)
```

```python
print("Polynomial Coefficients: ", poly_regressor.coef_)
print("Polynomial Intercept: ", poly_regressor.intercept_)
```

```python
print("Variance Score of Polynomial Regression: ", poly_regressor.score(X_poly_test, y_test))
```

The variance score or R^2 is best on degree=2 which is 0.9... it gives negative value when degree=3,4,5...


```python
print("MSE:", metrics.mean_squared_error(y_test, y_poly_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_poly_pred)))
print("MAE:", metrics.mean_absolute_error(y_test, y_poly_pred))
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(poly_regressor.predict(X_poly), poly_regressor.predict(X_poly) - y_train,
            color="green", s=10, label="Train data")
plt.scatter(poly_regressor.predict(X_poly_test), poly_regressor.predict(X_poly_test) - y_test,
            color="blue", s=10, label="Test data")

plt.hlines(y=0, xmin=min(y_test), xmax=max(y_test), linewidth=2)
plt.legend(loc="upper right")
plt.title("Polynomial Regression Residuals")
plt.xlabel("Predicted Profit")
plt.ylabel("Residuals (Error)")
plt.show()
```

```python
test_data = np.array([[1, 0, 0, 160000, 130000, 300000]])
test_data_poly = poly.transform(test_data)
predicted_profit = poly_regressor.predict(test_data_poly)

print("Predicted Profit for test data:", predicted_profit)
```
